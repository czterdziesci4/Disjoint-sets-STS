#### last_comment_user
czyj komentarz wskazaliśmy z identyfikatora jako ostatni dla danej gry? Kolejność wyznaczy dla Ciebie kolumna id_comment, a szukana wartość znajduje się w kolumnie user.

```python
q = """

    with last_comments as (
    
        
    select c.id_movie, g.id_game, g.game_name, MAX(c.id_comment) as last_comment_id, c.user
    
    from comments as c
        left join
        
        game_movies as gm 
        on c.id_movie = gm.id_movie
        left join games as g
        on gm.id_game = g.id_game
        
        
        group by g.id_game
    
    

            )
    
    select game_name,user from last_comments

"""
# do sprawdzenie czy należy podciąganć wszystkie id_game


last_comment_user = pysqldf(q) 


last_comment_user.to_csv(path_or_buf  = "last_comment_user.csv",  encoding = 'utf-8' )
last_comment_user
```

#### last_comment_user_count_of_comments
– ile mamy komentarzy przypisanych do tego “ostatniego” komentatora?

```python
# do sprawdzennia bo nadpisałem błędnie coś inengo
q = """

    with last_comments_user as (
        with last_comments as (

            select c.id_movie, g.id_game, g.game_name, MAX(c.id_comment) as last_comment_id, c.user

                from comments as c
                    left join

                    game_movies as gm 
                    on c.id_movie = gm.id_movie
                    left join games as g
                    on gm.id_game = g.id_game


                    group by g.id_game

                    )

    select id_game,user from last_comments
        )
    
    select lcu.user, count(c.id_comment) as last_comment_user_count_of_comments 
    from last_comments_user as lcu
    join comments as c
    where lcu.user = c.user
    group by lcu.id_game
        
"""
last_comment_user_count_of_comments = pysqldf(q)

last_comment_user_count_of_comments.to_csv(path_or_buf  = "last_comment_user_count_of_comments.csv",  encoding = 'utf-8' )
last_comment_user_count_of_comments
```

#### count_of_detected_single_modes_per_movie
– tu prosimy Cię o informację dla ilu filmów istnieje pojedyncza dominanta w zbiorze osób komentujących dany film pod względem liczby komentarzy. Inaczej – szukamy osób, które były najaktywniejsze pod danym filmem i zliczamy te filmy, dla których mamy jednego lidera :)

```python
q = """

/* I simply count total comments per movie and user */

with user_comments_per_movie as (
            select 
                id_movie, 
                user, 
                count(*) as totat_comments

                from comments
                group by id_movie, user
                order by id_movie Asc, totat_comments DESC
                )


,

/* Than I make a rank of the numbers of comments per user and movie */

user_comments_ranks as (
            select id_movie, 
                   user,
                   totat_comments,
                   RANK () OVER (PARTITION BY id_movie ORDER BY totat_comments DESC) number1
                   

                from user_comments_per_movie


                order by id_movie, number1

                )

/* I choose only number1 and dominant users */
    
,

/* here is the list of users that are single dominantes per movie */
dominants_per_movie as (         
    select id_movie, user, totat_comments, number1, count(*) as dominant
    
    from user_comments_ranks
    
    where number1 = 1
    
    
    group by id_movie
    having dominant = 1

    )
    
    
    /* Simply count them */
    select count(*) from dominants_per_movie 


"""



count_of_detected_single_modes_per_movie = pysqldf(q)
count_of_detected_single_modes_per_movie
```

### count_movies_with_game_name_in_title
– na koniec chcemy poznać liczbę filmów dla gry, których tytuł zawiera nazwę samej gry. Nie ma znaczenia wielkość liter, nie akceptujemy literówek, ignorujemy nazwy, w których zamiast dokładnej formy wskazanej w games.csv jest wersja odmieniona przez przypadki :)

```python
q = """
    
    select  
            g.game_name,
            count(gm.id_movie)
            
           
           
               from games as g
                   left join game_movies as gm
                   on g.id_game = gm.id_game
            
            where lower(gm.title) like lower('%' || g.game_name ||'%')  
            group by g.game_name
            
            
            
"""


count_movies_with_game_name_in_title = pysqldf(q)
count_movies_with_game_name_in_title.to_csv(path_or_buf  = "count_movies_with_game_name_in_title.csv",  encoding = 'utf-8' )
count_movies_with_game_name_in_title
```
